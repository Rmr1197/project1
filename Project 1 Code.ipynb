# Axial Base Deep Profiler - Winter

First importing the data from the CSV file


```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt

dataframe = pd.read_csv("axial base deep winter.csv") 
```

Designating and calculating variable values

```python
z = dataframe.pressure
T = dataframe.temp
cond = dataframe.conductivity_millisiemens
S = dataframe.practical_salinity
ssp = 1449.2 + (4.6*T) -0.055*T**2+0.00029*T**3+(1.34 -0.01*T)*(S-35)+0.016*z
```

Finding the Number of Dives

```python
Time = dataframe.time
plt.scatter(Time,z, s=0.5)
plt.ylabel("Depth", fontsize=13)
plt.xlabel("Time", fontsize=13)
plt.title("Graphic for Number of Dives", fontsize=18)
```

```python
def getnumberofdives(depth):
    bottom = max(depth)
    atbottom = False
    divecount = 0
    for i in range(0, len(depth)):
        if(depth[i] > 0.95 * bottom and not atbottom):
            atbottom = True
        if(depth[i] < 0.95 * bottom and atbottom):
            atbottom = False
            divecount +=1
    return divecount
getnumberofdives(z)
numdives = getnumberofdives(z) 
divestext = "%s dives were performed"%(numdives)
print (divestext)
```

Plotting speed of sound profile (SSP) (Depth versus Speed of Sound)

```python
avg=np.polyfit(z,ssp,11)
avgssp=np.poly1d(avg)(z)
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.scatter(ssp,z, s=0.5, label='Winter');
ax1.scatter(avgssp,z, s=.2, label='Average')
plt.gca().invert_yaxis();
plt.ylabel("Depth (m)", fontsize=13);
plt.xlabel("Speed of Sound (m/s)", fontsize=13);
plt.title("Speed of Sound Profile", fontsize=18);
plt.legend(loc='upper right');
```
